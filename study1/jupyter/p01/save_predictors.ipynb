```python
import numpy as np
import pandas as pd
```

```python
x = np.random.rand(1000).reshape([10,10,10])
```

```python
d = {}
for i in range(10):
    d[str(i)]=pd.DataFrame(x[:,:,i])
```

```python
np.save('d', d)
np.savez_compressed('d1', d=d)
```

```python
d_load = np.load('d.npy', allow_pickle = True)
d1_load = np.load('d1.npz', allow_pickle = True)
```
